```python
# Text Encoder Model Implementation

# Import Required Libraries
from transformers import AutoTokenizer, AutoModel

# Load the model and tokenizer
model = AutoModel.from_pretrained('Simonlee711/Clinical_ModernBERT')
tokenizer = AutoTokenizer.from_pretrained('Simonlee711/Clinical_ModernBERT')

# Sample text for encoding
text = "The patient was diagnosed with hypertension and prescribed medication."
inputs = tokenizer(text, return_tensors='pt')
outputs = model(**inputs)

# Display the last hidden state
print(outputs.last_hidden_state)
```

```python
# CLIP Model Implementation
import torch
import clip
from PIL import Image

device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

image = preprocess(Image.open("C:/Users/ishaa/Downloads/ventricles-14-02.jpg")).unsqueeze(0).to(device)
text = clip.tokenize(["a photo of a cat", "a photo of a dog"]).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)

    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

print("Label probs:", probs)  
```
